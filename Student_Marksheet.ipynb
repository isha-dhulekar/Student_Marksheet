```python
import pandas as pd

student=str(input("Enter student name:"))
course=str(input("Enter course name:"))
year=str(input("Enter year:"))
seat_number=str(input("Enter seat number:"))
registration_number=str(input("Enter REG. No. :"))

subjectCode = []
subjectName= []
externalMarks= []
total = 0

print()

numberOfSubjects = int(input('Enter Number of Subjects: '))

print()

for i in range(numberOfSubjects):
  subjectCode.append(int(input('Enter Subject Code: ')))
  subjectName.append(input('Enter Subject Name '))
  tempMarks = int(input('Enter Internal Marks '))
  total = total + tempMarks 
  externalMarks.append(tempMarks)    
  print()

def calGrade(marks):
  if(marks >= 45 and marks <= 49):
      return 'D(pass)'
  if(marks >= 50 and marks <= 54):
      return 'B'
  if(marks >= 55 and marks <= 59 ):
      return 'B'
  if(marks >= 60 and marks <= 74 ):
      return 'A'
  if(marks >= 75 and marks <= 89 ):
      return 'A+'
  if(marks >= 90 and marks <= 100) :
      return 'O'
  return 'F(fail)'

df = pd.DataFrame(columns = ['Subject Code', 'Subject Name', 'External Marks', 'Grade'])

for i in range(numberOfSubjects):
    df.loc[i] = [subjectCode[i], subjectName[i], externalMarks[i], calGrade(externalMarks[i]) ]

print("               SURYADATTA COLLEGE OF MANAGEMENT INFORMATION RESEARCH AND TECHNOLOGY    ")
print("                       STATEMENT OF GRADE FOR : BSc_CS_2019 CREDIT PATTERN             ")
print("                                      SEM 4                                                 ")
print()
print("NAME:",student)
print("SEAT NUMBER:",seat_number)
print("REG.NO.:", registration_number)
print()

print(df)
print()

percentage=((total)/(numberOfSubjects*100))*100
Sgpa=(percentage + 7.5)/10  

print("Sgpa :", Sgpa)
```
